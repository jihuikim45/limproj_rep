# async, sync

```python
import asyncio
import time

async def invoke_async(llm, messages):
    response = await llm.ainvoke(messages)
    print(response.content)

async def invoke_parallel(llm, messages):
    tasks = [invoke_async(llm, messages) for _ in range(10)]
    await asyncio.gather(*tasks)
```

```python
messages = [
    ("system","당신은 서울의 음식과 문화 전문가입니다."),
    ("human","서울을 대표하는 음식을 맛볼 수 있는 레스토랑 5개를 추천해주세요.")
]
start = time.perf_counter()

await invoke_parallel(llm, messages)

end = time.perf_counter()


print(f' Elased time : {end - start:.2f} second')
```

```python
# sync 호출 시간 비교 5times

import asyncio
import time

async def invoke_async(llm, messages1):
    response = await llm.ainvoke(messages1)
    print(response.content)

async def invoke_parallel(llm, messages1):
    tasks = [invoke_async(llm, messages1) for _ in range(5)]
    await asyncio.gather(*tasks)
```

```python
messages1 = [
    ("system","당신은 서울의 음식과 문화 전문가입니다."),
    ("human","서울을 대표하는 음식을 맛볼 수 있는 레스토랑 5개를 추천해주세요.")
]
start = time.perf_counter()

await invoke_parallel(llm, messages1)

end = time.perf_counter()


print(f' Elased time : {end - start:.2f} second')
```
